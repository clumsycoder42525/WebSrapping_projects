```python
import requests
from bs4 import BeautifulSoup

url = "https://www.meesho.com/women-men-kids-footwear/clp/15mz"
data = requests.get(url)
# Change 'html-parser' to 'html.parser'
soup = BeautifulSoup(data.content, "html.parser")
# Pass data.content instead of just data to BeautifulSoup
soup
```

```python
name=[]
rating=[]
price=[]
aprice=[]
disc=[]
deli= []
```

```python
for nm in soup.find_all("p",class_="sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU"):
    name.append(nm.text)
name
```

```python
for rt in soup.find_all("span",class_="Rating__StyledPill-sc-12htng8-1 cgkleE"):
    rating.append(rt.text)

rating

```

```python
for p in soup.find_all("h5",class_="sc-eDvSVe dwCrSh"):
  price.append(p.text)
price
```

```python
for ap in soup.find_all("span",class_="sc-eDvSVe XndEO NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc"):
  aprice.append(ap.text)
aprice
```

```python
rating.extend(['4.2', '4.1', '4.1', '4.0', '4.0', '4.0', '4.0', '4.0', '4.1', '4.0', '4.1'])
```

```python
data={
    "Name":name,
    "Price":price,
    "Reviews":aprice,
    "Delivery":deli,
    "Rating":rating
}
print(len(name))
print(len(price))
print(len(aprice))
print(len(deli))
print(len(rating))
```

```python
aprice.append("300 Reviews")
```

```python
for de in soup.find_all("span",class_="sc-eDvSVe fkvMlU NewProductCardstyled__OfferedShippingDesktop-sc-6y2tys-15 bpZaGb NewProductCardstyled__OfferedShippingDesktop-sc-6y2tys-15 bpZaGb"):
  deli.append(de.text)
for dc in soup.find_all("span",class_="sc-eDvSVe fkvMlU"):
  deli.append(dc.text)
deli
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

```python
def deliv (value):
  if (value =="Free Delivery"):
    return 0
  else:
    ar=value.split()
    return ar[1]
```

```python
df["Delivery"]=df["Delivery"].map(deliv)
```
